2.Exploratory Data Analysis

Gender

```python
ax = sns.countplot(x='Gender', data=df, palette=["brown","navy"])
sns.set(rc={'figure.figsize':(6,5)})
plt.title("Gender's count")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
colors=["red","green"]
sns.barplot(x='Gender' , y = 'Amount', data = sales_gen, palette=colors)
sns.set(rc={'figure.figsize':(6,4)})
plt.title("Purchasing Power of Male and Female")
plt.gca().set_facecolor('pink')  # gca means 'get current axes'
plt.show()
```

summery : from above graph we can she that most of the buyers are females and even the purchasing power of female are greater than men.

Age Group -->

```python
ax = sns.countplot(data = df, x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(10,5)})
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_age
```

```python
colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan'] 
sns.barplot(x="Age Group", y = "Amount", data=sales_age, legend=False, hue=x)
plt.title("Total Amount Vs Age Group")
```

From above graph we can see that most of the buyers are of age group between 26-35 yrs female

Marital Status -->

```python
colors = ["red","yellow"]
mar_sts = sns.countplot(data=df, x="Marital_Status", palette=colors)

for bars in mar_sts.containers:
    mar_sts.bar_label(bars)
sns.set(rc={'figure.figsize':(4,5)})
```

```python
sales = df.groupby(["Marital_Status","Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sales
```

```python
sns.barplot(data=sales, x="Marital_Status", y="Amount", hue="Gender")
sns.set(rc={'figure.figsize':(4,5)})
```

From above bar we conclude that most of the buyers are married women.

Occupation -->

```python
sns.set(rc={'figure.figsize':(20,5)})
# colors=["red","green","blue","violet","pink","orange","blue","black","maroon","cyan","magenta","teal","brown","grey","lime"]
occupation = sns.countplot(data=df, x="Occupation", palette="Set3")

for bars in occupation.containers:
    occupation.bar_label(bars)
```

From above graph we analyse that most of the buyers are working in IT sector, Aviation and Healthcare sector.

Product Category-->

```python
sns.set(rc={'figure.figsize':(20,5)})
pd_cat = sns.countplot(data=df, x="Product_Category", palette="Set1")

for bars in pd_cat.containers:
    pd_cat.bar_label(bars)
```

```python
# Boxplot (for outlier detection in numerical features)
sns.boxplot(x=df['State'],y=df['Orders'], palette="Set2")
plt.title("Total orders from states")
sns.set(rc={'figure.figsize':(22,5)})
plt.show()
```

```python
# 3.Scatter plot (display individual data points)

x = df['State'].head(11)
y = df['Amount'].head(11)

plt.scatter(x,y,color='red')
plt.title('Total amount/sales from States')
plt.xlabel("States")
plt.ylabel("Amount")
# plt.figure(figsize=(10,10))
sns.set(rc={'figure.figsize':(10,6)})
# plt.gca().set_facecolor('pink')
plt.show()
```

from above graph we can see that most of the orders & Total sales/amount are from UP, Maharashtra,AP,karnatka respectively.

```python
# Heatmap 
print("Correlation Matrix")
corr = display(df.corr(numeric_only=True))
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='plasma',fmt='.4f',linewidth=0.2)
plt.title("Correlation Heatmap")
sns.set(rc={'figure.figsize':(10,5)})
plt.show()
```

```python
# Pairplot for numerical variables (for finding the pairwise relationship between the numerical variables)
sale_cols = ["Age","Amount","Orders"]
sns.pairplot(df[sale_cols])
plt.suptitle("pairwise relationship ", y=1.02)
sns.set(rc={'figure.figsize':(15,5)})
plt.show()
```
